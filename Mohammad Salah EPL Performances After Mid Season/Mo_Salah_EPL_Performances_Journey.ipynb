```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

# Data from TransferMarkt
data = {
    'Season': ['2017/18', '2018/19', '2019/20', '2020/21', '2021/22', '2022/23', '2023/24', '2024/25'],
    'Goals': [15, 12, 9, 13, 15, 7, 12, 17],
    'Assists': [3, 6, 4, 3, 9, 4, 7, 13]
}
df = pd.DataFrame(data)

# Create the plot
plt.figure(figsize=(14, 9))
plt.style.use('dark_background')  # Set dark background

# Plot goals and assists
plt.plot(df['Season'], df['Goals'], marker='o', color='red', label='Goals', linewidth=2)
plt.plot(df['Season'], df['Assists'], marker='o', color='gold', label='Assists', linewidth=2)

# Annotate data points with values
for i, (season, goal, assist) in enumerate(zip(df['Season'], df['Goals'], df['Assists'])):
    plt.text(i, goal + 0.5, str(goal), color='red', fontsize=15, ha='center',fontweight='bold')
    plt.text(i, assist - 0.75, str(assist), color='gold', fontsize=15, ha='center',fontweight='bold')

# Add title and labels
plt.title("Mo Salah's Premier League Performances After Half a Season(Liverpool): A Comparison Through the Years", fontsize=18, color='white', loc='center',fontweight='bold')
plt.xlabel("Season", fontsize=14, color='white',fontweight='bold')
plt.ylabel("Count", fontsize=14, color='white',fontweight='bold')

# Customize ticks
plt.xticks(color='white', fontsize=12,fontweight='bold')
plt.yticks(color='white', fontsize=12,fontweight='bold')

# Adjust y-axis limits to provide more space at the top
plt.ylim(0, max(df['Goals']) + 2)  # Increase the upper limit slightly (add 2)

# Add grid and legend
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.legend(loc='upper left', fontsize=12, facecolor='black', edgecolor='white', framealpha=0.9)

# Add Mohamed Salah's image
image_url = "https://backend.liverpoolfc.com/sites/default/files/styles/lg/public/2024-08/mohamed-salah-alternative-body-v2-2024.png?itok=VMLtOKVH"
response = requests.get(image_url)
img = Image.open(BytesIO(response.content))

# Display image on the right side of the plot
plt.figimage(img.resize((260, 330)), xo=1150, yo=63)
# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
