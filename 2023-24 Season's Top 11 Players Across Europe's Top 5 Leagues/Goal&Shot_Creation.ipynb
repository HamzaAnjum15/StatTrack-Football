```python
#Importing relevant package
import pandas as pd

#Retrieving data from top5 european league players
url = "https://fbref.com/en/comps/Big5/2023-2024/gca/players/2023-2024-Big-5-European-Leagues-Stats"

df = pd.read_html(url, attrs={"id": "stats_gca"})

#Getting the first item from the list
df = df[0]

#Displaying dataset
df.head()
```

```python
#Dropping per 90 stats and creating our own later
df = df[df.columns.drop(list(df.filter(regex='Per 90')))]

#Dropping top header that we dont need
df = df.droplevel(0, axis = 1)
```

```python
df.head()
```

```python
df['90s'] = pd.to_numeric(df['90s'], errors='coerce')  # Converts to float
df['90s'].dtypes
```

```python
df.head()
```

```python
df[['SCA', 'GCA', 'TO']] = df[['SCA', 'GCA', 'TO']].apply(pd.to_numeric, errors='coerce')
df[['SCA', 'GCA', 'TO']].dtypes
```

```python
df[['SCA', 'GCA', 'TO']] = df[['SCA', 'GCA', 'TO']].fillna(0).astype(int)
```

```python
df.head()
```

```python
#Calculating per 90 metrics
df['ShotsCreatingAction_per_90'] = df['SCA'] / df['90s']
df['GoalsCreatingAction_per_90'] = df['GCA'] / df['90s']
# df['TakeOnLeadToGoal_per_90'] = df['TO'] / df['90s']
```

```python
df.head()
```

```python
df['Shot-Creating Actions'] = (df['ShotsCreatingAction_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Goal-Creating Actions'] = (df['GoalsCreatingAction_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
```
