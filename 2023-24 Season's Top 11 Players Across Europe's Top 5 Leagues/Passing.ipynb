```python
#Importing relevant package
import pandas as pd

#Retrieving data from top5 european league players
url = "https://fbref.com/en/comps/Big5/2023-2024/passing/players/2023-2024-Big-5-European-Leagues-Stats"

df = pd.read_html(url, attrs={"id": "stats_passing"})

#Getting the first item from the list
df = df[0]

#Displaying dataset
df.head()
```

```python
#Dropping per 90 stats and creating our own later
df = df[df.columns.drop(list(df.filter(regex='Per 90')))]

#Dropping top header that we dont need
df = df.droplevel(0, axis = 1)
```

```python
df.head()
```

```python
df.columns
```

```python
df['90s'] = pd.to_numeric(df['90s'], errors='coerce')  # Converts to float
df['90s'].dtypes
```

```python
df[['KP', '1/3', 'PPA','CrsPA']] = df[['KP', '1/3', 'PPA','CrsPA']].apply(pd.to_numeric, errors='coerce')
df[['KP', '1/3', 'PPA','CrsPA']].dtypes
```

```python
df[['KP', '1/3', 'PPA','CrsPA']] = df[['KP', '1/3', 'PPA','CrsPA']].fillna(0).astype(int)
```

```python
df.head()
```

```python
#Calculating per 90 metrics
df['KeyPasses_per_90'] = df['KP'] / df['90s']
df['PassessIntoFinalThird_per_90'] = df['1/3'] / df['90s']
df['PassessIntoPenaltyArea_per_90'] = df['PPA'] / df['90s']
df['CrossesIntoPenaltyArea_90'] = df['CrsPA'] / df['90s']
```

```python
df['Key Passess'] = (df['KeyPasses_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Passes Into Final Third'] = (df['PassessIntoFinalThird_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Passess Into Penalty Area'] = (df['PassessIntoPenaltyArea_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Crosses Into Penalty Area'] = (df['CrossesIntoPenaltyArea_90'].rank(pct=True) * 100).fillna(0).astype(int)
```

```python
df.head()
```

```python
players = ['Erling Haaland','Florian Wirtz', 'Kylian Mbappé', 'Harry Kane', 'Lamine Yamal',
           'Rodri', 'Jude Bellingham', 'Vinicius Júnior', 'Dani Carvajal',
           'Lautaro Martínez', 'Toni Kroos']

columns_to_display = ['Player','Squad','Key Passess', 'Passes Into Final Third', 'Passess Into Penalty Area','Crosses Into Penalty Area']

filtered_columns_df = df[df['Player'].isin(players)][columns_to_display]
filtered_columns_df
```
