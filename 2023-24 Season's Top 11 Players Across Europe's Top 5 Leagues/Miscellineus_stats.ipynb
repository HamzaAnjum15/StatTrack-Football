```python
#Importing relevant package
import pandas as pd

#Retrieving data from top5 european league players
url = "https://fbref.com/en/comps/Big5/2023-2024/misc/players/2023-2024-Big-5-European-Leagues-Stats"

df = pd.read_html(url, attrs={"id": "stats_misc"})

#Getting the first item from the list
df = df[0]

#Displaying dataset
df.head()
```

```python
#Dropping per 90 stats and creating our own later
df = df[df.columns.drop(list(df.filter(regex='Per 90')))]

#Dropping top header that we dont need
df = df.droplevel(0, axis = 1)
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['90s'] = pd.to_numeric(df['90s'], errors='coerce')  # Converts to float
```

```python
df['90s'].dtypes
```

```python
df[['Won', 'Recov', 'TklW', 'Int']] = df[['Won', 'Recov', 'TklW', 'Int']].apply(pd.to_numeric, errors='coerce')
```

```python
df[['Won', 'Recov', 'TklW', 'Int']].dtypes
```

```python
df[['Won', 'Recov', 'TklW', 'Int']] = df[['Won', 'Recov', 'TklW', 'Int']].fillna(0).astype(int)
```

```python
df[['Won', 'Recov', 'TklW', 'Int']].dtypes
```

```python
#Calculating per 90 metrics
df['AerialWon_per_90'] = df['Won'] / df['90s']
df['BallRecovery_per_90'] = df['Recov'] / df['90s']
df['TacklesWon_per_90'] = df['TklW'] / df['90s']
df['Interceptions_per_90'] = df['Int'] / df['90s']
```

```python
df.head()
```

```python
df[df['Player'] == 'Mats Hummels']
```

```python
df['Aerial Wons'] = (df['AerialWon_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Ball Recovery'] = (df['BallRecovery_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Tackles Won'] = (df['TacklesWon_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Interceptions'] = (df['Interceptions_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
```

```python
players = ['Erling Haaland', 'Florian Wirtz','Kylian Mbappé', 'Harry Kane', 'Lamine Yamal',
           'Rodri', 'Jude Bellingham', 'Vinicius Júnior', 'Dani Carvajal',
           'Lautaro Martínez', 'Toni Kroos']

columns_to_display = ['Player','Squad','Aerial Wons', 'Ball Recovery', 'Tackles Won','Interceptions']

filtered_columns_df = df[df['Player'].isin(players)][columns_to_display]
filtered_columns_df
```
