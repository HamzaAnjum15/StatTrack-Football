```python
#Importing relevant package
import pandas as pd

#Retrieving data from top5 european league players
url = "https://fbref.com/en/comps/Big5/2023-2024/possession/players/2023-2024-Big-5-European-Leagues-Stats"

df = pd.read_html(url, attrs={"id": "stats_possession"})

#Getting the first item from the list
df = df[0]

#Displaying dataset
df.head()
```

```python
#Dropping per 90 stats and creating our own later
df = df[df.columns.drop(list(df.filter(regex='Per 90')))]

#Dropping top header that we dont need
df = df.droplevel(0, axis = 1)
```

```python
df['90s'] = pd.to_numeric(df['90s'], errors='coerce')  # Converts to float
df['90s'].dtypes
```

```python

df[['Att','Succ','Att Pen']] = df[['Att','Succ','Att Pen']].apply(pd.to_numeric, errors='coerce')
df[['Att','Succ','Att Pen']].dtypes
```

```python
df[['Att','Succ','Att Pen']] = df[['Att','Succ','Att Pen']].fillna(0).astype(int)
```

```python
df.head()
```

```python
#Calculating per 90 metrics
df['TakeOnAttempted_per_90'] = df['Att'] / df['90s']
df['SuccessfulTakeOn_per_90'] = df['Succ'] / df['90s']
df['TouchesInAttackingPenaltyArea_per_90'] = df['Att Pen'] / df['90s']
```

```python
df['Dribbles Attempted'] = (df['TakeOnAttempted_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Successful Dribbles'] = (df['SuccessfulTakeOn_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Touches In Attacking Penalty Area'] = (df['TouchesInAttackingPenaltyArea_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
```

```python
players = ['Erling Haaland', 'Florian Wirtz','Kylian Mbappé', 'Harry Kane', 'Lamine Yamal',
           'Rodri', 'Jude Bellingham', 'Vinicius Júnior', 'Dani Carvajal',
           'Lautaro Martínez', 'Toni Kroos']
```

```python
columns_to_display = ['Player','Squad','Dribbles Attempted', 'Successful Dribbles', 'Touches In Attacking Penalty Area']

filtered_columns_df = df[df['Player'].isin(players)][columns_to_display]
filtered_columns_df
```
