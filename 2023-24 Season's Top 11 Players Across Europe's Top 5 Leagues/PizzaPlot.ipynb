```python
from urllib.request import urlopen

import matplotlib.pyplot as plt
from PIL import Image

from mplsoccer import PyPizza, add_image, FontManager
```

```python
font_normal = FontManager('https://raw.githubusercontent.com/googlefonts/roboto/main/'
                          'src/hinted/Roboto-Regular.ttf')
font_italic = FontManager('https://raw.githubusercontent.com/googlefonts/roboto/main/'
                          'src/hinted/Roboto-Italic.ttf')
font_bold = FontManager('https://raw.githubusercontent.com/google/fonts/main/apache/robotoslab/'
                        'RobotoSlab[wght].ttf')
```

```python
# # URL = "https://www.mancity.com/meta/media/cfzdcpm2/rodrigo.png?width=600"
# URL = "/content/RodriCircle.png"

# lewa_cropped = Image.open(urlopen(URL))

from PIL import Image

# Local file path
URL = "/content/ToniKroosCircle.png"

# Open the image
lewa_cropped = Image.open(URL)
lewa_cropped.show()  # Optional: Displays the image
```

```python
# parameter list
# params = [
#     "Non-Penalty Goals", "npxG", "npxG per Shot", "xA", "Open Play\nShot Creating Actions",
#     "\nPenalty Area\nEntries", "Progressive Passes", "Progressive Carries",
#     "Successful Dribbles", "\nTouches\nper Turnover", "pAdj\nPress Regains", "Aerials Won"
# ]
params = [
    "Goals", "Assists","Expected Goals", "Expected Assists", "Goal-Creating\n Actions","Open Play\nShot-Creating\n Actions","Successful\n Dribbles",
    "Touches In\n Attacking \n Penalty Area","Passess Into \n Final Third", "Passess Into \n Penalty Area","Crossess Into \n Penalty Area","Key Passess","Progressive\n Passes", "Progressive\n Carries",
    "Aerials Won","Tackles \n Won", "Ball \n Recoveries"
]
# params = [
#     "Goals", "Assists","Expected \n Goals", "Expected \n Assists", "Goal-Creating\n Actions",
#     "Progressive \n Carries","Progressive \n Passess","Key Passess","Passess Into \n Final Third","Passess In \n Penalty Area","Crossess Into \n Penalty Area",
#     "Shots \n Blocked","Passess \n Blocked","Aerial \n Wons","Tackles \n Won", "Ball \n Recoveries", "Interceptions", "Clearances"
# ]
# Player	Squad	Goals	Assists	Goals + Assists	Progressive Carries	Progressive Passes	Progressive Receptions	Expected Goals	Expected Assists
# Harry Kane	Bayern Munich	99	91	99	48	53	68	99	89
# # values for corresponding parameters
# The values are taken from fbref
values = [54, 95, 54, 90, 87, 90, 30, 23, 99, 70, 70, 95, 99, 53, 23, 80, 82]
```

```python
# Instantiate PyPizza class
baker = PyPizza(
    params=params,                  # list of params
    straight_line_color="#FFFFFF",  # white color for straight lines
    straight_line_lw=1,             # linewidth for straight lines
    last_circle_lw=1,               # linewidth of last circle
    other_circle_lw=1,              # linewidth for other circles
    other_circle_ls="-.",           # linestyle for other circles
    inner_circle_size=20            # increase the circle size
)

# Plot pizza
fig, ax = baker.make_pizza(
    values,              # list of values
    figsize=(8, 8),      # adjust figsize according to your need
    param_location=110,  # where the parameters will be added
    kwargs_slices=dict(
        facecolor="#6caddf", edgecolor="#FFFFFF",  # red slices with white edges
        zorder=2, linewidth=1
    ),                   # values to be used when plotting slices
    kwargs_params=dict(
        color="#FFFFFF", fontsize=12,          # white text for parameters
        fontproperties=font_normal.prop, va="center"
    ),                   # values to be used when adding parameters
    kwargs_values=dict(
        color="#FFFFFF", fontsize=12,          # white text for values
        fontproperties=font_normal.prop, zorder=3,
        bbox=dict(
            edgecolor="#FFFFFF", facecolor="#6caddf",  # red background for values
            boxstyle="round,pad=0.2", lw=1
        )
    )                    # values to be used when adding parameter-values
)

# Set black background
fig.patch.set_facecolor("#000000")  # black background for the entire figure
ax.set_facecolor("#000000")         # black background for the plot area

# Add title
fig.text(
    0.515, 0.97, "Rodri - FC Manchester City", size=18,
    ha="center", fontproperties=font_bold.prop, color="#FFFFFF"  # white title
)

# Add subtitle
fig.text(
    0.515, 0.942,
    "Percentile Rank vs Top-Five League Center Defensive Midfielders | Season 2023-24",
    size=15, ha="center", fontproperties=font_bold.prop, color="#FFFFFF"  # white subtitle
)

# Add image (if applicable)
# Add image (adjusted positioning)
ax_image = add_image(
    lewa_cropped,
    fig,
    left=0.438,  # Adjusted for better centering
    bottom=0.43, # Adjusted for better alignment
    width=0.14,  # Slightly increased to fill space
    height=0.14  # Slightly increased to fill space
)
   # these values might differ when you are plotting

plt.savefig('RodriPercentile.png', dpi=300, bbox_inches='tight')

plt.show()
```
