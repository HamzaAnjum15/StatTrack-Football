```python
#Importing relevant package
import pandas as pd

#Retrieving data from top5 european league players
url = "https://fbref.com/en/comps/Big5/2023-2024/stats/players/2023-2024-Big-5-European-Leagues-Stats"

df = pd.read_html(url, attrs={"id": "stats_standard"})

#Getting the first item from the list
df = df[0]

#Displaying dataset
df.head()
```

```python
#Dropping per 90 stats and creating our own later
df = df[df.columns.drop(list(df.filter(regex='Per 90')))]

#Dropping top header that we dont need
df = df.droplevel(0, axis = 1)
```

```python
df.head()
```

```python
df.columns
```

```python
df[['90s', 'xG', 'xAG']] = df[['90s', 'xG', 'xAG']].fillna(0).apply(pd.to_numeric, errors='coerce')
```

```python
df[['Gls', 'Ast', 'G+A', 'PrgC', 'PrgP', 'PrgR']] = df[['Gls', 'Ast', 'G+A', 'PrgC', 'PrgP', 'PrgR']].apply(pd.to_numeric, errors='coerce')
```

```python
df[['Gls', 'Ast', 'G+A', 'PrgC', 'PrgP', 'PrgR']] = df[['Gls', 'Ast', 'G+A', 'PrgC', 'PrgP', 'PrgR']].fillna(0).astype(int)
```

```python
#Calculating per 90 metrics
df['goals_per_90'] = df['Gls'] / df['90s']
df['assists_per_90'] = df['Ast'] / df['90s']
df['goals_assists_per_90'] = df['G+A'] / df['90s']
df['progressive_carries_per_90'] = df['PrgC'] / df['90s']
df['progressive_passes_per_90'] = df['PrgP'] / df['90s']
df['progressive_receptions_per_90'] = df['PrgR'] / df['90s']
df['expected_goals_per_90'] = df['xG'] / df['90s']
df['expected_assists_per_90'] = df['xAG'] / df['90s']
```

```python
#Percentiles calculation
df['Goals'] = (df['goals_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Assists'] = (df['assists_per_90'].rank(pct=True) * 100).fillna(0).astype(int)
df['Goals + Assists'] = (df['goals_assists_per_90'].fillna(0).rank(pct=True) * 100).astype(int)
df['Progressive Carries'] = (df['progressive_carries_per_90'].fillna(0).rank(pct=True) * 100).astype(int)
df['Progressive Passes'] = (df['progressive_passes_per_90'].fillna(0).rank(pct=True) * 100).astype(int)
df['Progressive Receptions'] = (df['progressive_receptions_per_90'].fillna(0).rank(pct=True) * 100).astype(int)
df['Expected Goals'] = (df['expected_goals_per_90'].fillna(0).rank(pct=True) * 100).astype(int)
df['Expected Assists'] = (df['expected_assists_per_90'].fillna(0).rank(pct=True) * 100).astype(int)

```

```python
players = ['Erling Haaland', 'Florian Wirtz','Kylian Mbappé', 'Harry Kane', 'Lamine Yamal',
           'Rodri', 'Jude Bellingham', 'Vinicius Júnior', 'Dani Carvajal',
           'Lautaro Martínez', 'Toni Kroos']

columns_to_display = ['Player','Squad','Goals', 'Assists', 'Goals + Assists','Progressive Carries','Progressive Passes','Progressive Receptions','Expected Goals','Expected Assists']

filtered_columns_df = df[df['Player'].isin(players)][columns_to_display]
filtered_columns_df
```
